```python
'''
Here I modelled the high and low salaries on categories excluding job_description.

Start by loading in the cleaned data.
'''

df = pd.read_csv('./seek_clean.csv', index_col = None)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['salary'] = np.nan
```

```python
med_nbi = 118146.60
med_nda = 121000
med_nde = 125000
med_nds = 120999.45
med_sbi = 125000
med_sda = 150000
med_sde = 135000
med_sds = 140000
```

```python
'''
I define that a high salary is one that is greater than the median value for the job category. Here I create the 
high or normal salary in a new column called salary.
'''

for index, row in df.iterrows():
    if row['jbd'] == 'Normal Data Scientist' and row['pay_total2'] > med_nds:
        df.salary[index] = 1
    elif row['jbd'] == 'Normal Business Intelligence' and row['pay_total2'] > med_nbi:
        df.salary[index] = 1
    elif row['jbd'] == 'Normal Data Engineer' and row['pay_total2'] > med_nde:
        df.salary[index] = 1
    elif row['jbd'] == 'Normal Data Analyst' and row['pay_total2'] > med_nda:
        df.salary[index] = 1  
    elif row['jbd'] == 'Senior Business Intelligence' and row['pay_total2'] > med_sbi:
        df.salary[index] = 1    
    elif row['jbd'] == 'Senior Data Engineer' and row['pay_total2'] > med_sde:
        df.salary[index] = 1
    elif row['jbd'] == 'Senior Data Analyst' and row['pay_total2'] > med_sda:
        df.salary[index] = 1    
    elif row['jbd'] == 'Senior Data Scientist' and row['pay_total2'] > med_sds:
        df.salary[index] = 1
    else:
        df.salary[index] = 0
```

```python
'''
Here I determine the baseline by calculating the distribution of high and normal salary.
'''

df.salary.value_counts(normalize=True)
```

```python
'''
Here set X (and dummy it), y, train and test split it.
'''

X = df[['advertiser','job_post_date', 'jbd']].copy()
```

```python
X = pd.get_dummies(X, drop_first=True, dtype=float)
```

```python
y = df['salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
'''
Model it with Logistic Regression.
'''

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_score = model.decision_function(X_test)
```

```python
'''
Gridsearch to optimise hyperparameters.
'''
logreg = LogisticRegression(solver='liblinear')
C_vals = [0.0001, 0.001, 0.01, 0.1, .15, .25, .275, .33, 0.5, .66, 0.75, 1.0, 2.5, 5.0, 10.0, 100.0, 1000.0]
penalties = ['l1','l2']

gs = GridSearchCV(logreg, {'penalty': penalties, 'C': C_vals}, verbose=False, cv=3)
gs.fit(X_train, y_train)
```

```python
'''
Check best params and estimator
'''

gs.best_params_
```

```python
gs.best_estimator_
```

```python
'''
Create confusion matrix do see where model is predicting false positives and negatives.
'''

conmat = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))
confusion = pd.DataFrame(conmat, index=['high_salary', 'low_salary'],
                         columns=['predict_highsalary','predict_lowsalary'])

print(confusion)

roc_auc_score(y_test, y_score)
```
