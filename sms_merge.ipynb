```
import pandas as pd
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
bs = pd.read_csv(io.BytesIO(uploaded['before_sms.csv']))
asm = pd.read_csv(io.BytesIO(uploaded['after_sms.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```
bs.head()
```

```
asm.head()
```

```
asm.columns = ['BP', 'Avg Bill after SMS']
```

```
asm.head()
```

```
df = pd.merge(bs, asm, how='left', left_on='BP', right_on='BP')
```

```
df.head()
```

```
df.to_csv('sms_joined.csv')
files.download('sms_joined.csv')
```
